**Handling Missing Values**

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0,'last review':df['last review'].min()},inplace=True)
```

```python
df.dropna(subset= ['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Droping Columns
df = df.drop(columns=['license','house_rules'],errors = 'ignore')
```

```python
df.head()
```

```python
#Remove dolloar sign and convert to fload

#  df['price'] = pd.to_numeric(
#     df['price'].replace('[\$,]', '', regex=True),
#     errors='coerce'
# )

# df['service fee'] = pd.to_numeric(
#     df['service fee'].replace('[\$,]', '', regex=True),
#     errors='coerce'
# )

df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

**Remove Duplicates**

```python
df.drop_duplicates(inplace=True)
```

**Visualization**

1. What is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color ='red')
plt.title('Distribution of Listing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows that Airbnb listing prices are spread across many different price ranges. There is no one price range where most listings are grouped. The KDE line also shows this spread clearly, meaning the dataset has listings with different price levels.

2. How are different room types distributed?

```python
df['room type']
```

```python
df['room type'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='room type',data=df,color="lightgreen", order=df['room type'].value_counts().index)
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

3. How are listings distributed across different neighborhoods?

```python
df['neighbourhood group']
```

```python
df['neighbourhood group'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group',data=df,color="lightpink" , order=df['neighbourhood group'].value_counts().index)
plt.title('Distribution of Listings Across Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.show()
```

4. What is the relationship between price and room type?

```python
df['price'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type',data=df,palette='Set1')
plt.title('Relationship between Price and Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.legend(title='Room Type')
plt.show()
```

5. How has the number of reviews changed over time?
